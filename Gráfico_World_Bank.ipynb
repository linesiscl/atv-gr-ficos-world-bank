##DATA SCIENCE & STATISTICAL COMPUTING : Gráfico com World bank

A análise refere-se ao crescimento populacional por ano no Brasil. Foram selecionados alguns anos (1968, 1975, 1985, 1995, 2005, 2015, 2020 e 2022) para o primeiro gráfico, em linha, do indicador populacional.


Depois, os gráficos seguem analisando o ano de 2020 na seguinte ordem:


*   Em barra, com os países com maior aumento
*   Histograma, com a distribuição e frequência de aumento
*   Em pizza, com a proporção de aumento nos países que mais o registraram
*   Scatter, com a relação do crescimento entre os anos 2000 e 2020





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv('The_World_Bank_Population_growth_annual_.csv')
```

```python
print(df.head())
```

```python
brasil = df[df['country_name'] == 'Brazil']

anos_selecionados = ['1968', '1975', '1985', '1995', '2005', '2015', '2020', '2022']

brasil = brasil.set_index('country_name').T.loc[anos_selecionados]

plt.figure(figsize=(10,6))
plt.plot(brasil.index, brasil['Brazil'], marker='o', color="plum")
plt.title('Indicador ao longo dos anos - Brasil')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.grid(True)
plt.show()
```

```python

ano_2020 = df[['country_name', '2020']]
ano_2020 = ano_2020.sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(ano_2020['country_name'], ano_2020['2020'], color="plum")
plt.title('Aumento em 2020 - Top 10 países')
plt.xlabel('País')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

```python
valores_2020 = df['2020']

plt.figure(figsize=(10,6))
plt.hist(valores_2020.dropna(), bins=20, edgecolor='k', color='plum')
plt.title('Distribuição dos valores do Indicador em 2020')
plt.xlabel('Valor do Indicador')
plt.ylabel('Frequência')
plt.show()
```

```python
ano_2020 = df[['country_name', '2020']].dropna()

ano_2020 = ano_2020.sort_values(by='2020', ascending=False).head(10)

top_10_valores = ano_2020['2020'].astype(float)
top_10_paises = ano_2020['country_name']

plt.figure(figsize=(10,6))
plt.pie(top_10_valores, labels=top_10_paises, autopct='%1.1f%%', startangle=140)
plt.title('Proporção do aumento em 2020 - Top 10 países')
plt.show()
```

```python
# Selecionar os dados para os anos de 2000 e 2020
scatter_data = df[['country_name', '2000', '2020']].dropna()

# Gerar o gráfico de dispersão
plt.figure(figsize=(10,6))
plt.scatter(scatter_data['2000'], scatter_data['2020'],c='plum')
plt.title('Relação entre os valores do Indicador em 2000 e 2020')
plt.xlabel('Valor do Indicador em 2000')
plt.ylabel('Valor do Indicador em 2020')
plt.grid(True)
plt.show()
```
